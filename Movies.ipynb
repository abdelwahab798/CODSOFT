<h5>Import Libraries adn data 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("IMDb Movies India.csv", encoding='ISO-8859-1')
df.head()
```

<h5> look in the data 

```python
df.shape
```

<h5> Data cleaning

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
df["Year"].unique()
```

```python
df["Year"]=df["Year"].apply(lambda x: str(x))
```

```python
df["Year"]=df["Year"].apply(lambda x: x.replace("("," "))
df["Year"]=df["Year"].apply(lambda x: x.replace(")"," "))
```

```python
df["Year"]=pd.to_datetime(df["Year"], errors='coerce')
```

```python
df["Duration"].unique()
```

```python
df["Duration"]=df["Duration"].apply(lambda x: str(x))
```

```python
df["Duration"]=df["Duration"].apply(lambda x: x.split("min")[0])
```

```python
df["Duration"]=pd.to_numeric(df["Duration"], errors='coerce')
```

```python
df["Genre"].unique()
```

```python
df["Genre"].head(10)
```

```python
df["Genre"]=df["Genre"].apply(lambda x: str(x))
df["Genre"].str.strip()
df["Genre"]=df["Genre"].apply(lambda x: x.split(",")[0])
```

```python
df["Genre"].unique()
```

```python
df["Votes"]=pd.to_numeric(df["Votes"], errors='coerce')
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

<h5> feature engineering

```python
df["Actor 11"] = df.groupby("Actor 1")["Rating"].transform("mean")
df["Actor 22"] = df.groupby("Actor 2")["Rating"].transform("mean")
df["Actor 33"] = df.groupby("Actor 3")["Rating"].transform("mean")
df["Director1"] = df.groupby("Director")["Rating"].transform("mean")
df["Genre1"] = df.groupby("Genre")["Rating"].transform("mean")
```

<h5> Make EDA

```python
df.hist(column=["Rating","Votes","Year","Duration"], bins=20, color="blue", edgecolor="black",figsize=(20,15))  
```

```python
sns.countplot(x="Genre", data=df,color="green")
plt.title("Genre Count")
plt.xticks(rotation=90)
```

```python
top_10_directors = df['Director'].value_counts(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_10_directors.values, y=top_10_directors.index,palette='dark')
plt.title('Top 10 Directors with Most Movie Involvements')
plt.xlabel('Number of Movies')
plt.ylabel('Director')
```

```python
top_10_ACTOR=df['Actor 1'].value_counts(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_10_ACTOR.values, y=top_10_ACTOR.index, palette='dark')
plt.title('Top 10 Actors 1 with Most Movie Involvements')
plt.xlabel('Number of Movies')
plt.ylabel('Actor 1')
```

```python
avg_year=df.groupby("Year")["Rating"].mean().reset_index()
plt.figure(figsize=(15, 5))
sns.lineplot(x=avg_year["Year"], y=avg_year["Rating"])
plt.title('Average Rating by Year')
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
```

<h5> Select the coloums , make a model and spilt the data

```python
x=df[["Votes","Duration","Actor 11","Actor 22","Actor 33","Director1","Genre1"]]
y=df["Rating"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

<h5> the accuracy

```python
from sklearn.linear_model import LinearRegression 
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=model.score(x_test,y_test)
accuracy*100
```
